### 4. Testing for stationarity

```python
# The adfuller() function will import from the model from statsmodels for the test; however, running it will only return 
# an array of numbers. This is why you need to also define a function that prints the correct output from that array.

from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(data_sub['Value'])
```

In this case using a significance level of 5%, look at critical value (5%)-2.89 in this example.  This is smaller (not much) than the text statistic of -2.38, meaning that  the null hypothesis can't be rejected.  Meaning that there is a unit root in the data and the data is non-stationary. 

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(data_sub)
plt.savefig("Autocorrelations")
plt.show()
```

Vertical lines represent the lags in the series, with the blue area representing the confidence interval.  The lines that go above the blue edge are lags significantly correlated with each other. There are 4 lines above the confidence interval (blue area), indicating lags correlated with each other meaning that there is autocorrelated data, confirming that the data set is likely non-stationary(has a trend), supporting the Dickey-Fuller test conducted earlier. 
Can't forecast with non-stationary data so need to make it stationary by removing any unnecessary variance. 

### 5. Stationarizing the Industrial Production Durable Goods: Ship and boat Building data

```python
# Differencing - calculate the difference btwn every value (t) and 
# its lag (t-1), store new results in a new dataframe

data_diff = data_sub - data_sub.shift(1) # The df.shift(1) function turns the observation to t-1, making the whole thing t - (t -1)
```

```python
data_diff.dropna(inplace = True) # Here, you remove the missing values that came about as a result of the differencing. 
# You need to remove these or you won't be able to run the Dickey-Fuller test.
```

```python
data_diff.head()
```

```python
data_diff.columns
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(data_diff)
```

```python
dickey_fuller(data_diff)
```

The Critical values are less than the Test Statistic , and a satisfactory p-value <0.05 so data is now stationary and null hypothesis can be rejected.

```python
# Recheck autocorrelations
plot_acf(data_diff)
plt.savefig("Autocorrelation2")
plt.show()
```

The autocorrelation plot has improved, but there are still a few lines (3) above the confidence interval (blue area), meaning that the data stationary is weak.  But this meets the " unspoken rule of thumb" , that you need at least 10 significant lags (10 vertical lines above the confidence interval).
