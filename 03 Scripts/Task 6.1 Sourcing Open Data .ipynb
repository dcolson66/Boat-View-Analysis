Address columns where price and currency together, convert the price to Euro's name EUR_price, drop Currency and Price

Location column should be spit to Country, City and cleaned up garbage to "other" then drop Location and Other


```python
# Split Price column into seperate columns: 'Currency','Price'
df[['Currency','Price']]= df['Price'].str.split(' ', 1, expand=True)
```

```python
df.head(5)
```

```python
# Checking the types of currency
df['Currency'].value_counts()
```

```python
# Use current Euro exchange rate as of 3.2023 for EUR_Price then drop Currency 
def eurprice(Price,Currency):
    if Currency=='EUR':
        return Price
    elif Currency=='CHF':
        return int(Price) * 1.00
    elif Currency=='Â£':
        return int(Price) * 0.13
    elif Currency=='DKK':
        return int(Price) * 1.13
df['EUR_Price'] = df.apply(lambda x: eurprice(x.Price, x.Currency), axis=1)
```

```python
df.head(5)
```

```python
# Split Location - Country, City, Other 
df[['Country','City']]= df['Location'].str.split('Â»', 1, expand=True)
df[['City','Other']]= df['City'].str.split('Â»', 1, expand=True)
df['City']=df['City'].str.replace('¶','',regex=True)
df['City']=df['City'].str.replace('Ã','',regex=True)
```

```python
# Renaming column 'Number of views last 7 days'
df= df.rename(columns={'Number of views last 7 days': 'Views last 7 days'})
```

```python
# Drop redundent columns
df = df.drop(['Price','Location','Currency','Other'],axis=1)
```

```python
# Dropping the row(s) where 'Year Built' is 0
df = df[df['Year Built']!=0]
```
